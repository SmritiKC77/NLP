We can see that there are html tags,stop words, special characters . So we first remove them before feeding data to algorithm because they will not contribute to analyze the sentiment.

Text Preprocessing

1. Remove html tags
2. Lowercasing
3. Removing Stop Words


```python
df = df.sample(10000)
```

```python
df.shape
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['sentiment']= label.fit_transform(df['sentiment'])
```

```python
df.head()
```

```python
# Removing html tags
import re 
def remove_tags(raw_text):
    cleaned_text = re.sub(re.compile('<.*?>'),"",raw_text)
    return cleaned_text
```

```python
df['review'] = df['review'].apply(remove_tags)
df['review'][800]
```

Coverting to lowercase


```python
df['review'] = df['review'].apply(lambda x:x.lower())
df['review'][800]
```

```python
# Remove Stop words
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
from nltk.corpus import stopwords

sw_list = stopwords.words('english')
df['review'] = df['review'].apply(lambda x: [item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
df
```

# Training the model

```python
X = df.iloc[:,0:1]

y = df['sentiment']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 2)
```

```python
X_train.shape
```

```python
X_test.shape
```

1. Bag of words is used for vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()
```

```python
X_train_bow.shape
```

```python
X_test_bow.shape
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train_bow,y_train)
```

```python
y_predict = gnb.predict(X_test_bow)
```

Evaluation


```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score =accuracy_score(y_test,y_predict)
print(accuracy_score)
```

My accuracy is not that great because I have applied algorithm in only 10000 sample data.
