```python
import json

with open("intents.json") as json_file:
    data = json.load(json_file)

# Extract intents
intents = data["intents"]

documents = []
metadatas = []
ids = []

for intent in intents:
    tag = intent["tag"]
    patterns = intent["patterns"]

    documents.extend(patterns)


    metadata = {"tags": tag}  # Metadata containing the corresponding tag
    metadatas.extend([metadata] * len(patterns))


    pattern_ids = [f"{tag}_{i}" for i in range(len(patterns))]
    ids.extend(pattern_ids)



```

```python
import chromadb
from chromadb.utils import embedding_functions
chroma_client = chromadb.Client()
```

```python
!pip install sentence_transformers
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name= "all-mpnet-base-v2")
collection = chroma_client.create_collection(name="my_collection",embedding_function=sentence_transformer_ef)
```

```python
collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids #MiniLM lm to do the embeddings
)
```

```python
collection = chroma_client.delete_collection(name="my_collection")
```

```python
results = collection.query(
    query_texts=["what hour is it"],
    n_results=5,
    include= ['distances','metadatas','embeddings','documents'])

results
```
